```python
import pandas as pd
import matplotlib.pyplot as plb
data= {
    'StudentID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Maths': [85, 78, 92, 88, 76, 95, 89, 84, 91, 73],
    'Physics': [80, 85, 88, 90, 92, 81, 77, 95, 78, 83],
    'Chemistry': [78, 82, 84, 79, 81, 76, 85, 88, 92, 79],
    'Biology': [90, 75, 85, 88, 84, 89, 76, 91, 80 ,12]
}

df=pd.DataFrame(data)
x=df["Maths"]
y=df["Chemistry"]
plb.figure(figsize=(10,5))
plb.subplot(1,2,1)
plb.scatter(x,y,color='blue')
plb.title("scatter plot between maths and physics")
plb.xlabel('Maths SCORES')
plb.ylabel("Chemistry SCORES")


plb.subplot(1,2,2)
plb.plot(x,y,color='pink',linestyle="-",marker="o")
plb.title("line plot between maths and physics")
plb.xlabel('Maths')
plb.ylabel("Chemistry")
plb.tight_layout
plb.show()

```

```python
import plotly.express as px
data= {
    'Country':['united states','canada','brazil','russia','india'],
    'Values':[100,50,80,90,70]

}

fig = px.choropleth(
    data,
    locations ='Country',
    locationmode = 'country names',
    color = 'Values',
    color_continuous_scale ='Blues',
    title = 'choropleth map of values by country')
fig.show()

```
